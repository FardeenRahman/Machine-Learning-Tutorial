```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
file = pd.read_csv("homeprices.csv")
file
```

```python
%matplotlib inline 
plt.xlabel("Area(square feet)")
plt.ylabel("Price(USD)")
plt.scatter(file.area, file.price, marker = '+', c = 'r')
```

```python


reg = linear_model.LinearRegression()
reg.fit(file[["area"]], file.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123 * 3300 + 180616.43835616432
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
prc = reg.predict(d)
```

```python
d['price'] = prc
```

```python
d.to_csv("prediction.csv", index = False)
```

```python
%matplotlib inline 
plt.xlabel("Area(square feet)")
plt.ylabel("Price(USD)")
plt.scatter(file.area, file.price, marker = '+', c = 'r')
plt.plot(file.area, reg.predict(file[['area']]), marker = '+')
```
